```python
import pandas as pd

jobdataset = pd.read_csv('linkedindata.csv')
jobsdata = jobdataset['Jobsdescriptions'].str.strip().values.tolist()
jobsdata
```

```python
import re

def extract_contact_info(text):
    # Regular expressions for matching phone numbers and email addresses
    phone_regex = r'\b\d{3}[-.]?\d{3}[-.]?\d{4}\b'
    email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b'

    # Find all matches of phone numbers and email addresses in the text
    phone_numbers = re.findall(phone_regex, text)
    email_addresses = re.findall(email_regex, text)

    return phone_numbers, email_addresses

# Example usage
# text_list = [
#     'Role - Python DeveloperLocation - Houston, TX (ONSITE)Type - CONTRACT/FULLTIMEJob Description10 Years + of overall software development experience; PythonKnowledge and experience in Python, Django, Angular JS, Oracle highly desirable.Knowledge of version and revision control practices and proceduresWorking experience with database environment - understanding relational database structure and hands-on SQL knowledge to extract/manipulate data.At least 1+ years of experience in an Investment Bank (or similar) as a DeveloperSolid application design, development, and testing work experience.Strong sense of ownership, ability to work independently, and have strong communication skills.',
#     # Add more text elements here if needed
# ]
text_list = jobsdata
all_phone_numbers = []
all_email_addresses = []

for text in text_list:
    phone_numbers, email_addresses = extract_contact_info(text)
    all_phone_numbers.extend(phone_numbers)
    all_email_addresses.extend(email_addresses)

# Print the extracted phone numbers and email addresses
print("Phone Numbers:")
for phone_number in all_phone_numbers:
    print(phone_number)

print("\nEmail Addresses:")
for email_address in all_email_addresses:
    print(email_address)
```

```python
import csv
from itertools import zip_longest

# phone_numbers = ['732-593-8453', '732-231-6194', '732-593-8453', '571-622-3980']
# descriptions = ['Prashant@stellentit.comGtalk', 'Prashant@stellentit.comhttps', 'Sourabh@extendinfosys.comJob', 'Sourabh@extendinfosys.comAddress', 'Varma@hermitageinfotech.com', '732-289-6103varma@hermitageinfotech.comwww.hermitageinfotech.com']
phone_numbers = all_phone_numbers
descriptions = all_email_addresses
# Fill missing values with empty strings
filled_values = zip_longest(phone_numbers, descriptions, fillvalue='')

# Write the filled values to a CSV file
with open('output.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Phone Number', 'Description'])
    writer.writerows(filled_values)

```
